```python
import numpy as np
from numpy import *
import random
from random import sample
import matplotlib.pyplot as plt
```

```python
data = np.random.normal(0,10,150)
count, bins, ignored = plt.hist(data, 30, density=True)
plt.plot(bins,1/(10*np.sqrt(2*np.pi))*np.exp(-(bins-0)**2/(2*10**2)), linewidth=2, color='lime')
plt.show()
print(data)
```

```python
mat_X = np.zeros(shape=(150,3))

for i in range(0,150):
    for j in range(0,3):
        mat_X[i][j] = data[i]**j
        
print(mat_X)
```

```python
def sampleTrue(length):
    true_val = np.random.uniform(0,1,size=length)
    return true_val

true_val = sampleTrue(len(mat_X[0]))

print(true_val)
```

```python
y_data = np.zeros(shape=(150,1))
for i in range(0,150):
    y_data[i] = true_val[0]*mat_X[i][0] + true_val[1]*mat_X[i][1] + true_val[2]*mat_X[i][2]
    
print(y_data)

noise = np.random.normal(0,8,150)
print(noise)

for i in range(0,150):
     y_data[i] += noise[i]

    
print(y_data)
```

```python
plt.scatter(data,y_data, color = 'cyan')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
random_index = random.sample(range(0,150),150)

train_data = np.zeros(shape=(90,2))
for i in range(0,90):
    train_data[i,0] = data[random_index[i]] 
    train_data[i,1] = y_data[random_index[i]]


valid_data = np.zeros(shape=(30,2))
for i in range(90,120):
    valid_data[i-90,0] = data[random_index[i]]
    valid_data[i-90,1] = y_data[random_index[i]]

test_data = np.zeros(shape=(30,2))
for i in range(120,150):
    test_data[i-120,0] = data[random_index[i]]
    test_data[i-120,1] = y_data[random_index[i]]


print(train_data)
```

```python
mul1 = np.dot(np.transpose(mat_X),mat_X)
mul1 = np.linalg.inv(mul1)
mul2 = np.dot(np.transpose(mat_X),y_data)

param_sol = np.dot(mul1,mul2)

print(param_sol)
```

```python
train_error = 0
for i in range(0,90):
    calc_approx = param_sol[0]+param_sol[1]*train_data[i][0]+param_sol[2]*train_data[i][0]**2
    train_error += (calc_approx-train_data[i][1])**2

train_error/=2
print("error using closed form solution : "+str(train_error))
```

```python
# def test_func(x,theta):
#     result = theta[0] +x*theta[1] + (x**2)*theta[2]
#     return result
def test_func(x,theta):
    result = 0
    for i in range(len(theta)):
        result += theta[i]*(x**i)
    return result
```

```python
#implementing gradient descent
x_0 = np.array([0.5,1.0,3.0])
alpha = 0.000003
tol = 1e-3

def poly_reg(x_0,alpha,train_data,tol,lamb):
    x_1 = x_0
    error = 10 #for convenience (just to enter the loop)
    errors = []
    h_step = 0
    
    while (error >= tol):

        partial_errors = []

        for i in range(0,len(train_data)):
            func_res = test_func(train_data[i][0],x_0)
            partial_errors += [(func_res-train_data[i][1])**2]
            x_0 = x_1.copy()


            for j in range(len(x_0)):
                x_1[j] = x_0[j] - alpha*((func_res-train_data[i][1])*(train_data[i][0]**j)+lamb*x_0[j])





        errors.append(np.sum(partial_errors))


        error = np.linalg.norm(x_1-x_0,2)    
        h_step+=1
    

    return(x_1,h_step,errors)
    
```

```python
x_1,iterations, errors = poly_reg(x_0,alpha,train_data,tol,0)

print("x_1:")
print(x_1)

print("true:")
print(true_val)
```

```python
train_error = 0
for i in range(0,90):
    calc_approx = x_1[0]+x_1[1]*train_data[i][0]+x_1[2]*train_data[i][0]**2
    train_error += (calc_approx-train_data[i][1])**2

train_error/=2
print("the error with GD is : "+str(train_error))

plt.plot(errors, color="orange")
plt.show()
```

```python
#creating the cubic design matrix
designMat = np.zeros(shape=(150,4))

for i in range(0,150):
     for j in range(0,4):
        
        designMat[i][j] = (data[i]**j)
        
print(designMat)
```

```python
#sample the true values
true_cubic = sampleTrue(len(designMat[0]))
print(true_cubic)
```

```python
#generating the new y values using cubic parameters
cubic_y_data = np.zeros(shape=(150,1))
for i in range(0,150):
    cubic_y_data[i] = true_cubic[0]*designMat[i][0] + true_cubic[1]*designMat[i][1] + true_cubic[2]*designMat[i][2] + true_cubic[3]*designMat[i][3]
    
print(cubic_y_data)

noise = np.random.normal(0,8,150)
print(noise)

for i in range(0,150):
     cubic_y_data[i] += noise[i]

    
print(cubic_y_data)
```

```python
#generating the data sets and their y values
random_index = random.sample(range(0,150),150)

train_data = np.zeros(shape=(90,2))
for i in range(0,90):
    train_data[i,0] = data[random_index[i]] 
    train_data[i,1] = cubic_y_data[random_index[i]]


valid_data = np.zeros(shape=(30,2))
for i in range(90,120):
    valid_data[i-90,0] = data[random_index[i]]
    valid_data[i-90,1] = cubic_y_data[random_index[i]]

test_data = np.zeros(shape=(30,2))
for i in range(120,150):
    test_data[i-120,0] = data[random_index[i]]
    test_data[i-120,1] = cubic_y_data[random_index[i]]


print(train_data)
```

```python
#using our model to generate parameters
theta_init = np.array([0.8,0.5,0.6,1.0])
alpha = 0.00000003
tol = 1e-3

theta_final,iterations, errors = poly_reg(theta_init,alpha,train_data,tol,0)
print(theta_final)

train_error = 0
for i in range(0,90):
    calc_approx = theta_final[0]+theta_final[1]*train_data[i][0]+theta_final[2]*train_data[i][0]**2+theta_final[3]*train_data[i][0]**3
    train_error += (calc_approx-train_data[i][1])**2

train_error/=2
print("error with cubic: "+ str(train_error))

```

```python
#This is the model after using gradient descent 
plt.scatter(data,y_data, color = 'purple')
plt.xlabel('X')
plt.ylabel('Y')



h_y = np.zeros(shape=(150,1))

for i in range(0,150):
        h_y[i] = theta_final[0]+theta_final[1]*train_x[i]+theta_final[2]*train_x[i]**2+theta_final[3]*train_x[i]**3
        

plt.plot(train_x,h_y,color = "lime")
plt.show()
#model has overfit the data
```
